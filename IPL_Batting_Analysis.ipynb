```python
batsman_score1=batsman_score.groupby(['batsman','Season'])['batsman_runs'].sum().reset_index()
batsman_score1
```

```python
batsman_score1.Season=batsman_score1.Season.str.extract('(\d+)')
batsman_score1
```

```python
most_runs_average_strikerate['Id']=range(0,most_runs_average_strikerate.shape[0])
most_runs_average_strikerate
```

```python
first=most_runs_average_strikerate.pop('Id')
most_runs_average_strikerate.insert(1,'Id',first)
```

```python
most_runs_average_strikerate
```

```python
batsman_score1.groupby(['batsman'])['batsman_runs'].sum().reset_index().sort_values(by=['batsman_runs'], ascending=False)
```

```python
final_dataset=most_runs_average_strikerate[['batsman','Id']].merge(batsman_score1,left_on='batsman',right_on='batsman',how='right')
final_dataset
```

```python
final_dataset['Season'] = final_dataset['Season'].astype(int)
```

```python
final_dataset
```

```python
x=final_dataset[['Id','Season']].values
y=final_dataset[['batsman_runs']].values
print(x)
```

```python
final_dataset.dtypes
```

```python
LR = LinearRegression()
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.10,random_state=0)
LR.fit(x_train , y_train)
y_pred=LR.predict(x_test)
y_test
```
